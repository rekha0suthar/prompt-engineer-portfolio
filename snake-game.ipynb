```python
import os
import openai
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file

openai.api_key = 'API_KEY'
```

```python
def get_completion(prompt, model='gpt-3.5-turbo'):
    messages=[{"role":"user", "content":prompt}]
    response=openai.ChatCompletion.create(
    model=model,
    messages=messages,
    temperature=0,
    )
    return response.choices[0].message["content"]

def get_completion_from_messages(messages, model='gpt-3.5-turbo', temperature=0):
    response=openai.ChatCompletion.create(model=model, messages=messages, temperature=temperature)
    return response.choices[0].message["content"]
```

```python
prompt = f''' You act as software developer. follow given instructions and build snake game. 
1. create 16 x 16 size box in which snake will move. 
2. create snake.
3. when snake hit the wall reset the game.
4. create four button to move snake in all directions like up, down, left and right.
5. create a ball which will appear at random location in box and when snake hit the ball it will appear again at random location.
6. increase score when snake hit ball
'''
```

```python
response = get_completion(prompt)
print(response)
```
